Return values-
-Sends a value back to where the function was called
-stops the function immediately
-makes the function reusable

```python
#print vs return

def add(a,b):
    print(a+b)

result=add(5,3)
print(result)

#print is to display
#return is to give the value back
```

```python
def add(a,b):
    return a+b

result=add(5,3)
print(result)
```

```python
#return single value
def square(num):
    return num*num

print(square(4))

#return multiple values
def calculate(a,b):
    return a+b,a-b

result=calculate(10,4)
print(result)

sum_val,sub_val=calculate(10,5)
print(sum_val)
print(sub_val)
```

```python
#return with condition
def check_even(num):
    if num%2==0:
        return "even"
    else:
        return "odd"


print(check_even(7))
```

```python
#function stops when it executes return

def find_first_even(nums):
    for n in nums:
        if n%2==0:
            return n
            print("hey")  #this will not be executed since return has been executed
    return None


print(find_first_even([10,12,13,19]))
print(find_first_even([11,17,13,19]))
```

```python
#return list

def get_squares(nums):
    result=[]
    for n in nums:
        result.append(n*n)
    return result

print(get_squares([2,5,9]))
```

```python
#return boolean

def is_positive(num):
    return num>0

print(is_positive(-5))
```

```python
#return dictionary
def create_student(name,age):
    return {"name":name,"age":age}

print(create_student("Sameeha",22))
```

Docstring-
 A string written inside a function to explain what it does

 it helps:
 other developers
 future you
 interviewers
 documentation tools

 it uses triple quotes- """  """   


```python
def add(a,b):
    """
    Adds two numbers.
    """
    return a+b

add(7,6)
print(add.__doc__)  #to access the docstring written inside
```

```python
#google style docstring
def add(a,b):
    """
    Add two numbers.

    Args:
        a(int):First number
        b(int):Second number
    
    Returns:
        int:Sum of a and b
    """
    return a+b
```

```python
#numpy style docstring

def add(a,b):
    """
    Add two numbers.

    Parameters
    ----------
    a : int
        First number.
    b : int
        Second number.

    Returns
    -------
    int
        Sum of a and b
    """
    return a+b

```

```python
#*args- variable positional arguments
#*args collects extra positional arguments into a tuple
#variable non key-word arguments

#sum all numbers

def total(*numbers):
    total_sum=0
    for n in numbers:
        total_sum+=n
    return total_sum

print(total(10,20,40))
print(total(50,60))
```

```python
def greet(greeting,*names):
    for name in names:
        print(greeting,name)

greet("Good morning","Ram","Sam","Cam")
```

```python
def show_types(*args):
    for value in args:
        print(value,"->",type(value))

show_types(10,"Sam",5.5)
```

```python
#kwargs
#variable keyword arguments 

def show_details(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)

show_details(name="Sam",age=21)
```

```python
def check_email(**kwargs):
    if "email" in kwargs:
        return "Email provided"
    return "Email missing"

print(check_email(name="Rahul",email="test@gmail.com"))
```

```python
#update default settings
def settings(**red):
    config={"theme":"light","font":12}
    config.update(red)
    return config

print(settings(theme="dark"))
```

```python
#combined example

def example(a,*args,**kwargs):
    print("a:",a)
    print("args",args)
    print("kwargs",kwargs)

example(1,2,3,4,name="Manish",song="retro")
```

lambda(anonymous function)-
lambda= small function written in one line

syntax:
lambda parameters: expression

no return keyword needed.

use lambda for-

short logic
inside map filter
one-time use

```python
#normal function

def square(x):
    return x*x

#lambda function (single parameter)

square= lambda x:x*x
print(square(5))

#lambda function (multiple parameters)

add=lambda a,b:a+b
print(add(5,3))
```

```python
#check even

is_even=lambda x:x%2==0
print(is_even(6))

#get last character

last_char=lambda word:word[-1]
print(last_char("Python"))

#convert to uppercase

upper=lambda word:word.upper()
print(upper("data"))
```

```python
#map()
#applies function to every element

#normal loop

nums=[1,2,3]
squares=[]

for n in nums:
    squares.append(n*n)


#using map()

nums=[1,2,3]
squares=list(map(lambda x:x*x,nums))
print(squares)
```

```python
#add 10 to each

nums=[5,10,15]
result=list(map(lambda x:x+10,nums))
print(result)

#uppercase names

names=["amit","sanjana"]
result=list(map(lambda x:x.upper(),names))
print(result)
```

```python
#filter function
#returns only elements that satisfy condition

#normal loop
nums=[1,2,3,4]
evens=[]

for n in nums:
    if n%2==0:
        evens.append(n)
print(evens)

#using filter()

nums=[1,2,3,4]
evens=list(filter(lambda x:x%2==0,nums))
print(evens)
```

```python
#remove empty strings

data=["Rahul","","Saketh","","Soumya"]
result=list(filter(lambda x:x!="",data))
print(result)
```

```python
#reduce
#used for cumulative calculation
#have to import for this
```

```python
#sum of all elements

from functools import reduce
nums=[1,2,3,4]
result=reduce(lambda x,y:x+y,nums)
print(result)
```

```python
#concatenate strings

words=["Data","Science","is","a","subject"]
print(reduce(lambda x,y:x+" "+y,words))
```

```python
#find max

nums=[10,50,20]
print(reduce(lambda x,y:x if x>y else y,nums))
```

#SCOPE- LEGB RULE

scope determines where the variable lives

LEGB stands for:

L-LOCAL
E-ENCLOSING
G-GLOBAL
B-BUILT-IN

```python
#local scope
def add():
    a=5
    b=6
    print(a+b)

add()

#here a,b are local variables and they exist only inside function
```

```python
#global scope
age=24   

def show():
    print(age)

show()
#here age is the global variable
#function can read global variable
```

```python
#enclosing scope- nested functions

def outer():
    x=10
    def inner():
        print(x)
    inner()

outer()
#here x is enclosing scope variable
#inner function uses variable from outer function
```

```python
#built-in scope
k="names"
print(k)
print(len(k))
```

```python
len=6
print(len)

a="mango"
print(len(a))   #now len won't work because the value of len is overwritten
```

```python
del len   #deleting the int object variable 
print(len("mango"))
```

```python
#global + local same name
x=100

def show():
    x=50
    print("inside function",x)

show()
print("outside function",x)
#local variable does not change or affect global variable
```

```python
#changing global variable inside function

total=0

def add_amount(amount):
    global total
    total+=amount

add_amount(3000)
add_amount(500)
print(total)
```
