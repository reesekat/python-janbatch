When you open a file pythron creates:

1. a connection to the file
2. a file object in memory(RAM)
3. a system resource handle

if you don't close it that connection stays open

if you leave it open:

resources get wasted
system may slow down
you may not be able to open new files

File modes-

r READ
w WRITE (overwrites file)
a APPEND (adds content)

```python
#checking file mode

file=open("data.txt","r")
print(file.mode)
file.close()
```

```python
#context manager
#writing using with

with open("data.txt","w") as f:
    f.write("\nPython is easy")

#no need for f.close(), file closes automatically
```

```python
with open("data.txt","r") as f:
    content=f.read()

print(f.closed)
```

```python
#exceptions(error handling)
#try-except

try:
    num=int(input("enter a number:"))
except ValueError:
    print("invalid input")
```

ValueError - wrong value
TypeError - wrong data type
NameError - variable not defined
ZeroDivisionError - divide by zero error
FileNotFoundError - file missing
IndexError - List index wrong
KeyError - Dictionary key missing
AttributeError - Method doesn't exist
ImportError - module cannot be imported
Exception - catches almost all errors (not recommended for specific error handling)

```python
try:
    print(age)
except NameError:
    print("Variable not defined")
```

```python
try:
    print(10/0)
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    open("abc.txt","r")
except FileNotFoundError:
    print("File not found")
```

```python
try:
    numbers=[1,2,3]
    print(numbers[5])
except IndexError:
    print("Index out of range")
```

```python
try:
    data={"name":"Sam"}
    print(data["age"])
except KeyError:
    print("Key not found")
```

```python
#Exception is generic, it catches almost all errors
try:
    x=10/0
except Exception:
    print("Something went wrong")

#not recommended for specific errors
```

```python
#catching multiple errors
try:
    a=int(input())
    b=int(input())
    print(a/b)
except (ValueError,ZeroDivisionError):
    print("Invalid input")
```

Finally-

Finally is a block that always runs whether there is an error or not

Basic Syntax:
try:
    #risky code
except:
    #handle error
finally:
    #always runs

we use finally to:
1.close files
2.Close database connections
3.Release resources
4.cleanup actions

```python
try:
    print(10/0)
except ZeroDivisionError:
    print("Cannot divide by zero")
finally:
    print("done")
```

```python
#file handling example

file=None

try:
    file=open("tool.txt","r")
    print(file.read())
except FileNotFoundError:
    print("file not found")
finally:
    if file:
        file.close()
        print("file closed")
```

```python
#with return statement

def test():
    try:
        return "try block"
    finally:
        print("finally block")

print(test())
```

```python
try:
    print("try block")
except:
    print("error")
else:
    print("no error")
finally:
    print("finally block")
```
