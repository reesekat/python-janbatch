Python list vs numpy Array

numbers=[1,2,3,4]
data=[1,"hello",3.5]

-can store different datatypes
-stored in scattered memory
-slower for larger calculations

numpy array:
arr=np.array([1,2,3,4])

-stores same datatype
-stored in continuous memory
-much faster
-supports mathematical operations


Memory difference:

python list                              NumPy Array
-----------                             -------------
-flexible                             -fixed data type
-slower                               -faster
-different types allowed              -same type only
-memory scattered                     -memory is continuous

#BROADCASTING

broadcasting allows operations between arrays of different shapes

broad casting rules-
Two arrays are compatible if:

1. They have same shape
    or

2. One of them is 1
    or

3. Smaller array can stretch to match larger


```python
arr=np.array([1,2,3])

print(arr+5)
```

```python
#broadcasting 1D with 2D

arr=np.array([[1,2,3],
             [4,5,6]])

b=np.array([10,20,30])

print(arr+b)
```

```python
a=np.array(([1,2,5],
           [4,5,6]))

b=np.array([1,2])
print(a+b)

#broadcasting fails
```

```python
#column broadcasting

arr=np.array([[1,2,3],
              [4,5,6]])

col=np.array([[10],
             [20]])
print(arr+col)
```

#Vectorization
doing operations on entire array at once instead of using loops

Vectorization makes it:
-faster
-cleaner code
-less memory usage

```python
#square all numbers
arr=np.array([1,2,3,4])
print(arr**2)
```

```python
a=np.array([1,2,3])
b=np.array([4,5,6])

print(a+b)
```

```python
#conditional vectorization

arr=np.array([10,20,30,40])
print(arr>25)
```

```python
#using condition to modify values

arr[arr>25]=100
print(arr)
```

```python
#using where

np.where(arr>25,1,0)
```

```python
#mathematical operations

arr=np.array([1,5,3,4])

print(np.sort(arr))
print(np.sqrt(arr))
print(np.log(arr))
print(np.sin(arr))
```

```python
#broadcasting +vectorization together

arr=np.array([[1,2,3],
              [4,5,6]])

mean=np.array([2,3,4])

print(arr-mean)

#broadcasting handles shape, vectorization handles operation
```
