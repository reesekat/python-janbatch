```python
import pandas as pd
import numpy as np

data={
    'Customer_name':['Asha','Bala','Chitra','Deepa','Esha','Bala','Bala','Bala'],
    'City':['Bangalore','Delhi','Mumbai','Delhi','Bangalore','Delhi','Delhi','Delhi'],
    'Order_value':[5000,7000,np.nan,6500,7200,7000,7000,7000],
    'Email':['asha123@gmail.com','bala@gmail.com','chitra@company.com',
             'deepa_98@yahoo.com','esha@gmail.com','bala@gmail.com',
             'bala@gmail.com','bala@gmail.com']
}

df=pd.DataFrame(data)
df
```

Handling duplicate values

Duplicates occur when the same record appears multiple times in a dataset.

Common reasons:
Data entry errors
Merging datasets
System logging errors

Duplicates can distort the analysis,for example:
Revenue may appear higher than reality

```python
#identifying duplicates
df.duplicated()
```

```python
#viewing duplicate rows
df[df.duplicated()]
```

```python
#removes completely identical rows
df.drop_duplicates(inplace=True)
df
```

```python
#using subset

df1=df.drop_duplicates(subset='Email')
df1
#this removes duplicates only based on the Email column
```

```python
#keep decides which duplicate record remains
df1=df.drop_duplicates(keep=False)
df1

#keep parameter accepts only three values-first,last,false
```

Common regex patterns

\d     digit
\w     letter or number
^      start of a string
$      end of string
.      any character
+      one or more
*      zero or more


commonly used with -
.str.contains()
.str.extract()
.str.replace()
.str.match()

```python
import pandas as pd

data={"Email":["asha123@gmail.com",
              "bala@gmail.com",
              "chitra@company.com,"
              "deepa_98@yahoo.com"]}
df=pd.DataFrame(data)
```

```python
#extract numbers from text

df["Email"].str.extract(r"(\d+)")
```

```python
#find rows containing numbers
df["Email"].str.contains(r"\d")
```

```python
#extract email domain
df["Email"].str.extract(r"@(\w+)")
```

when we write @(\w+)

()       capture group
\w       letter or number
+        one or more

here match starts at the symbol @ but since @ is outside of the capture group it is not extracted
so only what is inside the capture group() is extracted

```python
#Find emails ending with gmail.com

df["Email"].str.contains(r"@gmail\.com$")
```

```python
data={"Name":["Asha","Bala","Chitra","Deepa"]}
df=pd.DataFrame(data)
```

```python
#Find names starting with A
df["Name"].str.contains(r"^A")
```

```python
#remove numbers from emails
df["Email"].str.replace(r"\d","",regex=True)

```

```python
#Regex groups
#deepa_98@gmail.com
#we want to extract both name and number in the email separately

df["Email"].str.extract(r"(\w+)_(\d+)")
```
