Pandas is a python library used for:

-reading data(csv,excel,etc)
-cleaning data
-analyzing data
-preparing data for visualisation

```python
import pandas as pd
```

```python
#series vs dataframe

#a series is like a single column of data
#it is one dimensional
```

```python
s=pd.Series([10,20,30,40])
print(s)
```

```python
#custom index

s=pd.Series([85,90,88],index=["Asha","Ravi","Kiran"])
print(s)
```

```python
#Dataframe
#a dataframe is a table (rows+columns)
#it is 2-dimensional

data={
    "Name":["Asha","Ravi","Kiran"],
    "Marks":[85,98,67]
}
df=pd.DataFrame(data)
print(df)

print(df.index)
print(df.columns)
print(df.values)
```
