f-string formatting types-

variable {x}
expression {x+y}
precision {x:.2f}
width {x:8}
alignment {x:^8}
padding {x:*^8}
zero padding {x:05}
percentage {x:.2%}
commas {x:,}

```python
print(f"The sum of the numbers is {a+b}") #expression

price=3600.78965
print(f"{price:.2f}")  #precision

x=42
print(f"{x:7}")  #width

p="python"
print(f"|{p:<15}|")  #left alignment

print(f"|{p:>15}|")  #right alignment

print(f"|{p:^15}|")  #center alignment

print(f"{42:*<6}")  #padding
print(f"{42:*>6}")
print(f"{42:*^6}")

print(f"{7:03}")    #zero padding
print(f"{123:06}")

value=3.5               #zero padding+precision
print(f"{value:07.2f}")

ratio=0.7865
print(f"{ratio:.2%}")   #percentage formatting

salary=12500000                 #commas(thousands separator)
print(f"{salary:,}")
print(f"{salary:,.2f}")

```

ESCAPE SEQUENCE CHARACTERS-

\'	Single Quote	
\\	Backslash	
\n	New Line	
\r	Carriage Return	
\t	Tab	
\b	Backspace

```python

print("This is also\na string")         #\n
print('This is his book and his book\'s color is red')   #\'
print("This is\ra string")  #\r carriage return
print("second string\\")   #backslash
print("this is \t tab string") #tab
print("This is backspace string\b")  #backspace
```
