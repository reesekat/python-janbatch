Loops

Loops are used to execute the same block of code multiple times without writing the code repeatedly.

There are two types-
FOR LOOP
WHILE LOOP

FOR LOOP-
A for loop runs once for every item in a collection

collection=list,string,dictionary,etc

syntax-
for item in iterable:
    action

```python
students=["Raj","Sam","Max","Rekha"]

for name in students:
    print(name)
```

```python
genders=["M","F","m"]

for g in genders:
    print(g.lower())
```

```python
#with condition
marks=[30,50,90]

for m in marks:
    if m>=40:
        print("pass")
    else:
        print("fail")
```

```python
word="DATA"
for w in word:
    print(w)
```

```python
#checking for vowels
word="daTA"
for ch in word:
    if ch.lower() in "aeiou":
        print(ch)
```

```python
words="data"
for c in words:
    if c in "aeiou":
        print(c)
```

```python
emp={"name":"rahul","dept":"IT"}

for i in emp:
    print(i)

for v in emp.values():
    print(v)

for k,v in emp.items():
    print(k,v)
```

```python
#email validation
emails=["abc@gmail.com","","123@gmail.com",""]

for e in emails:
    if not e:
        print("Missing")
    else:
        print("Send mail to:",e)
```

for -> fixed collection
while -> run until condition becomes false

WHILE LOOP-

Syntax-

while condition:
    action

```python
#counting example
i=1

while i<=5:
    print(i)
    i+=1
```

```python
value=int(input("enter a value"))

while value:
    print("value exists",value)
    break
```

```python
while True:
    print("Running forever")
```

```python
#stop at first failure
marks=[70,80,30,90]

for m in marks:
    if m<40:
        print("Fail found")
        break

#break statement is to stop the loop or program at a certain point
#here break is used to stop at the first failure
```

```python
#continue

values=["Rahul","","Amit"]

for v in values:
    if not v:
        continue
    print(v)

#empty string is skipped
#continue is used to skip current iteration and go to next
```

```python
#pass 
#placeholder for future code

for n in range(3):
    if n==1:
        pass       #do nothing
    else:
        print(n)
```

WHILE LOOP

1. print 1 to 5

2. print 5 to 1

3. print even numbers uptil 10

```python
i=1
while i<=5:
    print(i)
    i+=1
```

```python
i=5
while i>=1:
    print(i)
    i-=1
```

```python
#print even numbers uptil 10
i=2

while i<=10:
    print(i)
    i+=2
```

```python
#print every number multiplied by 2 uptil 5

i=1

while i<=5:
    print(i*2)
    i+=1
```

```python
#sum of numbers from 1 to 5

i=1
total=0
while i<=5:
    total+=i
    i+=1
print(total)

```

```python
#count digits in a number

number=12345
count=0

while number>0:
    number=number//10
    count=count+1
print("Digits:",count)
```

```python
#keep asking password until correct password

password=""

while password!="admin":
    password=input("enter the password")
    
    if(password!="admin"):
        print("enter the right password")
print("access granted")
```

```python
#loop through a list using while

names=["Rahul","Amit","Priya"]
i=0
while i<len(names):
    print(names[i])
    i=i+1
```

```python
#stop when value found
# stop the loop when 20 is found
nums=[10,11,12,20,30,40]

i=0
while i<len(nums):
    if nums[i]==20:
        print("value found")
        break
    print("number found is:",nums[i])
    i=i+1
```

```python
#skip even numbers between 1 to 10
i=1
while i<=10:
    if i%2==0:
        i=i+1
        continue
    print(i)
    i=i+1

```

```python
#skip odd numbers between 1 to 10
i=1
while i<=10:
    if i%2!=0:
        i=i+1
        continue
    print(i)
    i=i+1
```

```python
#find maximum in a list

nums=[10,45,20,99,30]
max_val=nums[0]

i=0
while i<len(nums):
    if nums[i]>max_val:
        max_val=nums[i]
    i=i+1

print(max_val)
```

```python
#print squares upto 10
i=1
while i<=10:
    square=i*i
    print(f"number:{i},square {square}")
    i+=1
```

```python
#print largest even number
nums=[11,22,33,44,66]
i=0
max_val=0

while i<len(nums):
    if nums[i]%2==0 and nums[i]>max_val:
        max_val=nums[i]
    i=i+1
print("largest even number is:",max_val)
```
