match-case

match-value:
    case option1:
        action
    case option2:
        action
    case _:
        default

```python
role=input("enter your role")

match role:
    case "admin":
        print("Full access")
    case "manager":
        print("Limited access")
    case _:
        print("access to only view")
```

```python
#ticket status
status="P"

match status:
    case "P":
        print("Pending")
    case "C":
        print("Closed")
    case "R":
        print("Resolved")
    case "W":
        print("Waiting")
    case _:
        print("Unknown")
```

```python
city="Delhi"

match city:
    case "Delhi"|"Bangalore":
        print("Metro city")
    case _:
        print("Tier 2 city")
```
