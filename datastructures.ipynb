Lists-
heterogenous,mutable,ordered,duplicates are allowed

```python
a=[1,3,"apple",4.5,True,4+5j,"apple"]
print(a)
print(a[3])
print(a[0:3])

a[4]="Orange"
print(a)
```

```python
#append - to add elements into the list,sits at the end of the list
a=[1,2,3,4]
a.append(5)
print(a)
```

```python
#extend - to add multiple elements to the list

a=[1,2,3,4,5]
a.extend([6,5,6])
print(a)

b=[9,10,11]
b.extend(a)
print(b)
```

```python
#index

a=[1,2,3,4,5]
a.index(5)

#insert
a.insert(0,1)   #insert(index position,element to insert)
print(a)
```

```python
#pop
a.pop()
print(a)

a.pop(0)
print(a)
```

```python
#remove
a=[1,2,3,4,5]
a.remove(4)
print(a)
```

```python
#clear
a.clear()
print(a)
```

```python
#count
a=[1,2,3,4,4,7,8,9,10,11,34]
a.count(4)
```

```python
#sort
a=[1,2,3,85,23,10,11,7]
b=a.sort()
print(a)

```

```python
#copy

a=[1,2,3,4]
b=a.copy()
print(b)


a.append(5)
print(a)
print(b)

a=[1,2,3,4,[5,6]]
b=a.copy()
print(a)

b[4][0]=100
print(b)
print(a)
```

```python
a=[1,2,3,4]
b=a

print(a)
print(b)

a.append(5)
print(a)
print(b)
```

```python
import copy
a=[1,2,3,4,[5,6]]
b=copy.deepcopy(a)
print(b)

b[4][0]=100
print(b)
print(a)
```

TUPLES-

Tuples are ordered,heterogenous,unchangeable,allows duplicates

```python
a=(1,2,3,"red",1.5,4,2,3)
print(type(a))
```

```python
print(a[0])
print(a[0:3])
```

```python
#tuple methods-count,index

a=(3,5,7,6,6,3,5,3)
print(a.count(3))
print(a.index(7))

t=6,7,8       #tuple packing
print(type(t))

a,b,c=t     #tuple unpacking
print(a)
print(b)
print(c)
```

```python
#built-in functions that can be used with tuple
t=(3,4,6,7,8,3)
print(len(t))
print(min(t))
print(max(t))
print(sum(t))
```

SETS-

ITEMS ARE UNORDERED,HETEROGENOUS,ELEMENTS MUST BE UNIQUE,DUPLICATES ARE NOT ALLOWED, ELEMENTS MUST BE IMMUTABLE.

HASHABILITY-
ELEMENTS STORED WITHIN THE SET MUST BE IMMUTABLE WHICH MEANS HASHABILITY IS FOLLOWED
AN ELEMENT IS CALLED HASHABLE WHEN IT IS OF IMMUTABLE TYPE AND CAN BE STORED WITHIN SET.

```python
#defining a set
a={"monday",2,4,2.5,True,4+2j,2,4}
print(a)
print(type(a))

b=set(["tuesday",2,2.5,False,3+4j])   #set method
print(type(b))
```

```python
a=[1,2,2,1,4,5,"mONDAY","Monday"]
print(a)

b=set(a)
print(b)
```

SET OPERATIONS-

UNION
INTERSECTION
DIFFERENCE
SYMMETRIC DIFFERENCE

```python
a={12,13,14}
b={14,15,16}

print(a|b)  #union of two sets
print(a & b) #intersection of two sets
print(a-b) #difference (in a but not in b)
print(a^b) #symmetric difference (not common)
```

SET METHODS-

ADD
UPDATE
REMOVE
DISCARD
POP
CLEAR
ISSUBSET
ISSUPERSET
ISDISJOINT

```python
#add
s={1,2,3}
s.add(4)
print(s)

#update 
s={1,2,3}
s.update([2,4,6])
s.update({7,8,9})
s.update((10,11,12))
print(s)

#remove
s.remove(6)
print(s)

s.discard(13)
print(s)

s.pop()
print(s)

s.clear()
print(s)

s={1,2,3}
t={1,2,3,4,5,6}
u={4,5,6}
print(s.issubset(t))
print(t.issuperset(s))
print(s.isdisjoint(u))

b=s.copy()
print(b)

```

UPDATE OPERATIONS

DIFFERENCE_UPDATE
INTERSECTION_UPDATE
SYMMETRIC_DIFFERENCE_UPDATE

```python
a={1,2,3,4,5}
b={2,3}

a.difference_update(b)
print(a)

a={1,2,3,4,5}
b={2,3}
a.intersection_update(b)
print(a)

a={1,2,3}
b={2,3,4}
a.symmetric_difference_update(b)  #removes elements present in both sets
print(a)
```
