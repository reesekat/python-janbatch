Functions-

Functions are for reusability and organisation

without functions- messy code
with functions- clean,modular,professional code

A resuable block of code that perfoms a task
instead of rewriting the logic again and again, we define it once and reuse it.

syntax-

def function_name():
    code

```python
#defining function without parameter
def greet():
    print("hello user,hope you are doing great")
```

```python
#calling function without argument
greet()
```

```python
#defining function with parameter
def greet1(name):
    print("Hello",name)
```

```python
#calling function with argument
greet1("rahul")
```

```python
def add(a,b):
    sum=a+b
    print(sum)
```

```python
add(1,2)
```
