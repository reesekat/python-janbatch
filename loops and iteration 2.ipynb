List comprehension
It is a shorter way to write a for loop that creates a list

syntax-
[new_value for item in list]

[new_value for item in list/iterable if condition]

```python
#square numbers

nums=[1,2,3,4]
squares=[]

for n in nums:
    squares.append(n*n)
print(squares)
```

```python
nums=[1,2,3,4]

squares=[n*n for n in nums]

print(squares)
```

```python
#convert to lowercase
names=["RAHUL","PRIYA","NIHAL"]
names_l=[n.lower() for n in names]
print(names_l)
```

```python
#in a new list filter only even numbers
nums=[1,2,3,4,5,6,7,8]
evens=[n for n in nums if n%2==0]
print(evens)
```

when to use list comprehension-

1. you are creating a list
2. you are transforming data
3. you are filtering items
4. logic is simple

when to not use-

1. when you are not building a list
2. when logic is complex
3. when you need break or continue
4. when updating multiple variables

Dictionary comprehension-
used to create dictionaries dynamically

syntax-
{key:value for item in iterable}

```python
#square mapping
nums=[1,2,3]
squares={n:n*n for n in nums}
print(squares)
```

```python
#print names along with their length
names=["Rahul","Priya"]
lengths={n:len(n) for n in names}
print(lengths)
```

```python
#mark students as pass/fail
marks={"Rahul":30,"Amit":60,"Priya":80,"Saketh":25}
status={name:"pass" if score>=40 else "fail" for name,score in marks.items()}
print(status)
```

```python
#keys in lowercase and values in upper case

data={"BOX":"red","HAMMER":"blue","SOCKS":"pink"}
new_data={k.lower():v.upper() for k,v in data.items()}
print(new_data)
```

NESTED LOOPS-
Loop inside another loop

-used for matrix,combinations,comparisons,grid logic

syntax/structure:

for x in list1:
    for y in list2:
        action

```python
#combinations

colors=["Red","Blue"]
sizes=["S","M"]

for c in colors:
    for s in sizes:
        print(c,s)
```

```python
#pattern printing

for i in range(1,4):
    for j in range(i):
        print("*",end="")
    print()
```

```python
for i in range(1,4):
    for j in range(1,4):
        print(i*j,end=" ")
    print()
```

```python
#matrix traversal

matrix=[[1,2],[3,4]]

for row in matrix:
    for value in row:
        print(value)
```

```python
#count elements in matrix
matrix=[[1,2],[3,4]]
count=0

for row in matrix:
    for value in row:
        count+=1
print("Total elements:",count)
```

```python
#find common names
students=["Rahul","Amit","Priya"]
present=["Amit","Priya"]

for name in students:
    if name in present:
        print(name)
```

```python
#number triangle

for i in range(1,4):
    for j in range(i):
        print(i,end=" ")
    print()
```

```python
#square of numbers

for i in range(1,4):
    for j in range(1,5):
        print(j,end=" ")
    print()
```
